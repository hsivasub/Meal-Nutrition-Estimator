# Health Score Engine
A rule-based and weighted scoring system to determine meal health ratings, establishing a score from 0 to 100 and mapping it to a Traffic Light label.

```python
import sys
import os
sys.path.append(os.path.abspath('..'))

from src.health_scorer import HealthScoreEngine
```

## 1. Initialize Scoring Engine

```python
engine = HealthScoreEngine()
```

## 2. Evaluate Differing Target Scenarios

```python
meals = {
    "Grilled Chicken Salad": {
        "calories": 350,
        "protein_g": 40,
        "fat_g": 12,
        "carbs_g": 15,
        "sugar_g": 4,
        "sodium_mg": 300
    },
    "Pasta Bolognese": {
        "calories": 650,
        "protein_g": 25,
        "fat_g": 22,
        "carbs_g": 75,
        "sugar_g": 8,
        "sodium_mg": 700
    },
    "Double Cheeseburger & Fries": {
        "calories": 1200,
        "protein_g": 45,
        "fat_g": 60,
        "carbs_g": 110,
        "sugar_g": 15,
        "sodium_mg": 1500
    }
}

for name, nutrition in meals.items():
    result = engine.evaluate_meal(nutrition)
    print(f"Meal: {name}")
    print(f"  -> Score: {result['health_score']}/100")
    print(f"  -> Label: {result['traffic_light']}\n")
```
