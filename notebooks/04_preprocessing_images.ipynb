```python
import sys
import os
sys.path.append(os.path.abspath('..'))

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from src.preprocessing import get_train_transforms, get_inverse_transforms
from src.data_loader import get_dataloaders
```

## 1. Test Transforms on a Single Sample Image

```python
# Assuming we have an image in data/sample/ (e.g. dummy test)
# Create dummy image if it doesn't exist
os.makedirs('../data/sample', exist_ok=True)
dummy_path = '../data/sample/dummy.jpg'
if not os.path.exists(dummy_path):
    Image.fromarray(np.random.randint(0, 256, (300, 300, 3), dtype=np.uint8)).save(dummy_path)

img = Image.open(dummy_path)
transform = get_train_transforms()
tensor_img = transform(img)

inv_transform = get_inverse_transforms()
display_img = inv_transform(tensor_img)

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(img)
ax[0].set_title('Original')
ax[1].imshow(display_img)
ax[1].set_title('Augmented & Transformed')
plt.show()
```
