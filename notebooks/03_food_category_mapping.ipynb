# Food-101 to USDA Mapping

In this notebook, we map the 20 MVP Food-101 categories to their closest matching USDA FoodData Central (`fdc_id`) entries. This lookup table acts as the bridge between our visual classifier and our nutritional database.

```python
import pandas as pd
import json
import os
from fuzzywuzzy import process
```

## 1. Load Food-101 MVP Classes
These are our selected top 20 food categories built into the MVP.

```python
mvp_classes = [
    'pizza', 'hamburger', 'sushi', 'french_fries', 'hot_dog',
    'fried_rice', 'ramen', 'ice_cream', 'donuts', 'macarons',
    'tacos', 'steak', 'spaghetti_bolognese', 'chicken_wings',
    'omelette', 'caesar_salad', 'dumplings', 'grilled_cheese_sandwich',
    'pancakes', 'waffles'
]

mvp_classes = [c.replace('_', ' ') for c in mvp_classes]
print("MVP Classes:", mvp_classes)
```

## 2. Text Search & Ambiguity Handling
If you search 'pizza', USDA returns hundreds of variations. For the MVP, we use exact or generic term lookups by writing simple heuristics or manually assigning the most common representative match.

```python
FOOD_CSV = "../data/usda/food.csv"

if os.path.exists(FOOD_CSV):
    food_df = pd.read_csv(FOOD_CSV)
    print(f"Loaded {len(food_df)} USDA food descriptions.")
    
    # Example of a fuzzy search logic
    # However, selecting the 'best' representative food logically requires either 
    # filtering by FDC category (e.g. "Survey (FNDDS)") or manual assignment.
    
    print("Matches for 'pizza':")
    matches = food_df[food_df['description'].str.contains('pizza', case=False, na=False)].head(5)
    display(matches[['fdc_id', 'description']])
else:
    print("USDA food.csv not found locally. For this notebook, we'll demonstrate static mapping construction.")
    
```
