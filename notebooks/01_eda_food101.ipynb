# Day 2: Food-101 Exploratory Data Analysis

This notebook loads the Food-101 dataset, explores class distributions, resolution statistics, and visualizes samples. Finally, it identifies 20 food classes to form our MVP.

```python
import os
import glob
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

%matplotlib inline
sns.set_theme(style="whitegrid")
```

## 1. Local Dataset Path
Assuming the Food-101 dataset from Kaggle has been downloaded and extracted to `../data/raw/`.

```python
DATA_DIR = "../data/raw/images"
if not os.path.exists(DATA_DIR):
    print(f"Dataset not found at {DATA_DIR}. Please download from Kaggle.")
```

## 2. Class Distribution
Let's count how many images exist per category.

```python
data = []
if os.path.exists(DATA_DIR):
    classes = os.listdir(DATA_DIR)
    for cls in classes:
        cls_path = os.path.join(DATA_DIR, cls)
        if os.path.isdir(cls_path):
            num_images = len(os.listdir(cls_path))
            data.append({"class": cls, "num_images": num_images})

    df = pd.DataFrame(data)
    print(f"Total classes: {len(df)}")
    print(f"Total images: {df['num_images'].sum()}")

    # Plot
    plt.figure(figsize=(15, 6))
    sns.histplot(df['num_images'], bins=20)
    plt.title('Distribution of Images per Class')
    plt.xlabel('Number of Images')
    plt.ylabel('Frequency (Classes)')
    plt.show()
else:
    print("Waiting for dataset...")
```

## 3. Sample Images

```python
if os.path.exists(DATA_DIR):
    sample_classes = random.sample(classes, 4)
    fig, axes = plt.subplots(1, 4, figsize=(16, 4))

    for ax, cls in zip(axes, sample_classes):
        img_name = random.choice(os.listdir(os.path.join(DATA_DIR, cls)))
        img_path = os.path.join(DATA_DIR, cls, img_name)
        img = Image.open(img_path)
        ax.imshow(img)
        ax.axis('off')
        ax.set_title(cls)
    plt.tight_layout()
    plt.show()
```

## 4. Resolution Statistics

```python
if os.path.exists(DATA_DIR):
    # Taking a random subset to compute stats quickly
    resolutions = []
    for cls in random.sample(classes, 10):  # select 10 classes
        for img_name in random.sample(os.listdir(os.path.join(DATA_DIR, cls)), 50):
            img_path = os.path.join(DATA_DIR, cls, img_name)
            with Image.open(img_path) as img:
                resolutions.append({'width': img.width, 'height': img.height})
    
    res_df = pd.DataFrame(resolutions)
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=res_df, x='width', y='height', alpha=0.5)
    plt.title('Image Resolutions (Sample)')
    plt.show()
```
